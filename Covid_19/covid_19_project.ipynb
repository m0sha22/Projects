# VISUALLIZING WORLD DATA


## Confirmed cases (Total Cases) all over the world

```python
df_world_over_time= world_df.groupby(['date'])[["total_cases","new_cases","total_deaths","new_deaths"]].sum().reset_index()
df_world_over_time.date = pd.to_datetime(df_world_over_time.date)
df_world_over_time.set_index('date', inplace=True)
df_world_over_time.total_cases
```

```python
# Using line-plot
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time.index, df_world_over_time.total_cases)
plt.title('Evolution of Confirmed Covid-19 cases over time in the word', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('Confirmed cases', fontsize=16)
plt.show()
```

```python
#Total deaths cases evolution over time
# Using line-plot
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time.index, df_world_over_time['total_deaths'])
plt.title('Evolution of Covid-19 Deaths cases over time in the world', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('Number of Deaths', fontsize=16)
```

```python
# new cases evolution over time
# Using line-plot
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time.index, df_world_over_time['new_cases'])
plt.title('Evolution of Covid-19 New cases over time in the world', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('Number of New Cases', fontsize=16)
```

```python
# Using line-plot
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time.index, df_world_over_time['total_cases'], label='Confirmed')
plt.plot(df_world_over_time.index, df_world_over_time['total_deaths'], label='Total Deaths')
plt.plot(df_world_over_time.index, df_world_over_time['new_cases'], label='New')
plt.plot(df_world_over_time.index, df_world_over_time['new_deaths'], label='New Deaths')
plt.legend(loc=0)
plt.show()
```

# Exploring the USA data

```python
usa_df_total = df[df['location'] == 'United States']
dict(usa_df_total.isna().sum())
```

```python
usa_df = usa_df_total[['date','total_cases', 'new_cases', 'total_deaths', 'new_deaths','female_smokers','male_smokers','median_age']]
usa_df.head()
```

```python
usa_df.info()
```

```python
# fillling nan values
usa_df['new_cases'].fillna(method = "bfill", inplace=True)
usa_df['total_deaths'].fillna(method ='bfill', inplace=True)
usa_df['total_cases'].fillna(method = "bfill", inplace=True)
usa_df['new_deaths'].fillna(method = 'bfill', inplace=True)
usa_df['female_smokers'].fillna(method = 'bfill', inplace=True)
usa_df['male_smokers'].fillna(method = 'bfill', inplace=True)
usa_df['date'].fillna(method = 'bfill', inplace=True)
```

```python
# Convert date to datetime
pd.to_datetime(usa_df.date)
```

```python
# set the index to the date
usa_df.info()
```

```python
# seeing the recent values
usa_df[usa_df.date == usa_df.date.max()]
```

```python
#reset the index
usa_df.reset_index(drop=True, inplace=True)
```

## Visualizing the data 

```python
plt.figure(figsize=(20,10))
plt.plot(usa_df.date, usa_df['new_cases'])
plt.xlabel('date')
plt.ylabel('New cases')
plt.title('USA New cases')
```

```python
# New cases in January 2021
jan_df_usa = usa_df[(usa_df['date'] >= '2021-01-01') & (usa_df['date'] < '2021-02-01')][['date', 'new_cases']]
plt.figure(figsize=(20,10))
plt.plot(jan_df_usa.date, jan_df_usa.new_cases)
plt.xlabel('January_2021')
plt.ylabel('New cases')
plt.title('USA New cases in January 2021')
```
